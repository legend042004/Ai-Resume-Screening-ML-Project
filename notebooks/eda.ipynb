```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

DATA = Path("../data/UpdatedResumeDataSet.csv")   # relative to /notebooks

df = pd.read_csv(DATA)
display(df.head())
print(df.info())
print(df['Category'].value_counts())

df['Category'].value_counts().plot(kind='bar', figsize=(10,4))
plt.title("Resume Categories")
plt.show()
```
